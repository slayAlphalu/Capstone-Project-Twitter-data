# Clean and transform LIWC dictionary from pdf converted excel

```python
import pandas as pd
```

```python
data = pd.read_excel('LIWC2015 dictionary poster-converted.xlsx', skiprows=[0,1])
```

```python
data.head()
```

## Change column names to be categories

```python
columns = list(data.columns)
```

```python
for i in range(len(columns)):
    columns[i] = columns[i].split('\n')[-1]

columns = columns[:-1]
columns
```

```python
data = data.iloc[:,:-1]
```

```python
data.columns = columns
```

```python
data.head()
```

## Clean and format words under each category

```python
data.iloc[0,0]
```

```python
for i in range(len(columns)):
    words = data.iloc[0, i]
    words = words.replace('\n', ' ')
    words = words.split()
    data.iloc[0, i] = " ".join(str(x) for x in words)
```

```python
data.head()
```

```python
data = data.transpose()
```

```python
data.head()
```

## Save the dictionary to a txt file

Each row in the txt file would start with a category name followed by a comma, and a string of words in that category seperated by white space. 

```python
data.to_csv('LIWC_dic.txt', header = False)
```
